```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
```

```python
X= np.array([[1],[2],[3],[4],[5]])
y=np.array([56,67,78,89,90])
model = LinearRegression()
model.fit(X,y)
houres= float(input("Enter hours you studied :"))
pridictedMarkks=model.predict([[houres]])
print(f'pridected marks of {houres} hours is {pridictedMarkks}')
```

```python
X= np.array([1,2,3,4,5])
Y=np.array([10,20,30,40,50])

sxy=np.sum(X*Y)
sx=np.sum(X)
sy=np.sum(Y)
length=len(X)
x1=np.sum(X*X)
x2=np.sum(X)*np.sum(X)
meanX=np.mean(X)
meanY=np.mean(Y)

m=(length*sxy-sx*sy)/(length*x1-x2)


b=meanY-m*meanX

x=int(input("enter valaue"))
y=m*x+b
print('Pridicted value',y)
```

```python
from sklearn.linear_model import LogisticRegression
X= np.array([[1],[2],[3],[4],[5]])
y=np.array([0,0,1,1,1])
model=LogisticRegression()

model.fit(X,y)
hours= float(input("Enter hours you studied :"))

result = model.predict([[hours]])[0]
if result==1:
    print("Pass")
else:
    print("Fail")
```

```python
from sklearn.neighbors import KNeighborsClassifier
x=[[180,7],[200,7.5],[250,8],[300,8.5],[330,9],[350,9.5]]
y=[0,0,0,1,1,1]
#0 apple, 1 orange
model= KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)

weight = float(input("Enter weight"))
size = float(input("Enter size"))

result=model.predict([[weight,size]])
if result==0:
    print('base on pridiction it is apple')
else:
    print('base on pridiction it is Orange')
```

```python
from sklearn.tree import DecisionTreeClassifier
x=[[3,1],[4,2],[5,3],[6,4],[7,5],[8,6]]
y=[0,0,0,1,1,1]

model=DecisionTreeClassifier()

model.fit(x,y)

size = float(input("Enter size"))
shade = float(input("Enter shade"))

result=model.predict([[size,shade]])

if result==0:
    print('it a apple')
else:
    print('it is Orange')
```
