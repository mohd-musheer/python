```python
class A:
    v=10
    def take2(self,x,y):
        print(x+y)
    def take3(self,x,y,z):
        print(x+y+z)


ob1 =A()
print(ob1.v)
ob1.take2(5,5)
ob1.take3(8,6,5)
```

```python
class B (A):
    g=10
    def take4(self,a,b,c,d):
        print(a+b+c+d)
        
ob2=B()
print(ob2.g)
ob2.take4(4,5,6,7)
print(ob2.v)
ob2.take2(1,2)
ob2.take3(6,5,7)
```

```python
import threading
def PrintCat():
    for i in range(10):
        print("Cat")
def PrintDog():
    for i in range(10):
        print("Dog")
        
t1=threading.Thread(target=PrintCat)
t2=threading.Thread(target=PrintDog)
t1.start()
t2.start()
t1.join()
t2.join()
```
